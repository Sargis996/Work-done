```python
# Turns out not pre_preprocessing (Tokenizer has default preprocessing) gives a better model
pre_preprocess = False 

# Function for pre_preprocessing reviews
def clean_reviews(reviews):
    if not pre_preprocess:
        return reviews
    
    clean = []
    stop_words = ['i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', "you're", "you've", "you'll", "you'd", 'your', 'yours', 'yourself', 'yourselves', 'he', 'him', 'his', 'himself', 'she', "she's", 'her', 'hers', 'herself', 'it', "it's", 'its', 'itself', 'they', 'them', 'their', 'theirs', 'themselves', 'what', 'which', 'who', 'whom', 'this', 'that', "that'll", 'these', 'those', 'am', 'is', 'are', 'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had', 'having', 'do', 'does', 'did', 'doing', 'a', 'an', 'the', 'and', 'but', 'if', 'or', 'because', 'as', 'until', 'while', 'of', 'at', 'by', 'for', 'with', 'about', 'against', 'between', 'into', 'through', 'during', 'before', 'after', 'above', 'below', 'to', 'from', 'up', 'down', 'in', 'out', 'on', 'off', 'over', 'under', 'again', 'further', 'then', 'once', 'here', 'there', 'when', 'where', 'why', 'how', 'all', 'any', 'both', 'each', 'few', 'more', 'most', 'other', 'some', 'such', 'no', 'nor', 'not', 'only', 'own', 'same', 'so', 'than', 'too', 'very', 's', 't', 'can', 'will', 'just', 'don', "don't", 'should', "should've", 'now', 'd', 'll', 'm', 'o', 're', 've', 'y', 'ain', 'aren', "aren't", 'couldn', "couldn't", 'didn', "didn't", 'doesn', "doesn't", 'hadn', "hadn't", 'hasn', "hasn't", 'haven', "haven't", 'isn', "isn't", 'ma', 'mightn', "mightn't", 'mustn', "mustn't", 'needn', "needn't", 'shan', "shan't", 'shouldn', "shouldn't", 'wasn', "wasn't", 'weren', "weren't", 'won', "won't", 'wouldn', "wouldn't"]
    # For each review 
    for review in reviews:
        # Remove special characters and numbers
        review = re.sub('[^A-Za-z]+', ' ', review)

        # Make review lower case
        review = review.lower()

        # Split into tokens
        review = review.split()
        
        # Remove stop words
        for word in review:
            if word in stop_words:
                review.remove(word)

        # Join the words to form the cleaned review
        clean.append(" ".join(review))
    
    return clean
```

```python
top_n_occurring_words = 51429
# Tokenize the text
samples = clean_reviews(train['review'])
tokenizer = Tokenizer(num_words = top_n_occurring_words)
tokenizer.fit_on_texts(samples)

max_length = 200
# Convert text to sequences
sequences = tokenizer.texts_to_sequences(samples)
data = pad_sequences(sequences, maxlen = max_length)
```

```python
# Print out the number of unique tokens
word_index = tokenizer.word_index
print(f'Found {len(word_index)} unique tokens.')

# Shape of data
print("Number of training review samples:", data.shape[0])
print("Number of tokens in each review:", data.shape[1])
Found 51429 unique tokens.
Number of training review samples: 161297
Number of tokens in each review: 200
```

```python
# Categorize ratings as required and one-hot encoded them to feed to model
# labels = 2.0, when ratings >= 7
# labels = 1.0, when ratings >= 5 and ratings < 7
# labels = 0.0, when ratings < 5

ratings = train['rating'].values
labels = 1.0*(ratings >= 7) + 1.0*(ratings >= 5)
labels = to_categorical(labels)
```

```python
# Split into training and validation datasets
train_data, val_data, train_labels, val_labels = train_test_split(data, labels, test_size = 0.2)
```

```python
from keras import regularizers

word_vector_dimension = 400

# Conv1D model
Conv1D_model = Sequential()
#Conv1D_model.add(Embedding(top_n_occurring_words, word_vector_dimension, input_length = max_length, embeddings_regularizer=regularizers.l2(0.01)))
Conv1D_model.add(Embedding(top_n_occurring_words, word_vector_dimension, input_length = max_length))
#Tried different Conv1D layers and how many of these layers should be used
#Conv1D_model.add(Conv1D(word_vector_dimension, 5, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
Conv1D_model.add(Conv1D(word_vector_dimension, 5, activation='relu'))
#Conv1D_model.add(Conv1D(word_vector_dimension, 5, activation='relu', strides=3))
#Conv1D_model.add(Conv1D(word_vector_dimension, 7, activation='relu'))
#Conv1D_model.add(Conv1D(word_vector_dimension, 7, activation='relu', strides = 3))
#Conv1D_model.add(Conv1D(word_vector_dimension, 9, activation='relu'))
Conv1D_model.add(MaxPooling1D(5))
#GlobalAveragePooling1D()
# We tried both of them with different amounts of dropout
Conv1D_model.add(LSTM(word_vector_dimension, dropout = 0.3, recurrent_dropout = 0.3)) # Recurrent nerual network
#Conv1D_model.add(LSTM(word_vector_dimension, dropout = 0.3, recurrent_dropout = 0.5)) # Recurrent nerual network
#Conv1D_model.add(GRU(word_vector_dimension, dropout = 0.3, recurrent_dropout = 0.5)) # A gating mechanism in recurrent neural networks
Conv1D_model.add(Dense(3, activation='softmax'))

Conv1D_model.summary()

Conv1D_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])
#Conv1D_model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['acc'])
```

```python
# Include callbacks to help find the optimal model

# Stops training when val_acc stops improving after 5 epochs
early_stopping = EarlyStopping(
    monitor='val_acc',
    patience=5,
    min_delta=1e-7,
    restore_best_weights=True,
    verbose=0
)

# Reduces learning rate when val_acc stops improving after 2 epochs
# learning_rate_reduction = ReduceLROnPlateau(
#     monitor='val_acc',
#     factor = 0.1,                                     
#     patience = 2,                                   
#     min_delt = 1e-7,                                                            
#     verbose = 0
# )
```

```python
Conv1D_epochs = 10 #Not all 10 epoch are rull all the time.
Conv1D_batch_size = 1024

# Train the Conv1D model
start = time.time()
Conv1D_hist = Conv1D_model.fit(train_data, 
                        train_labels, 
                        epochs=Conv1D_epochs, 
                        batch_size=Conv1D_batch_size, 
                        validation_data = (val_data, val_labels),
                        callbacks = [early_stopping])

end = time.time()

print(f'Time used: {end-start} seconds')
print('\n')
```

```python
# Accuracy plot
plt.title("Accuracy plot for Conv1D")
plt.plot(Conv1D_hist.history['acc'], label='Accuracy')
plt.plot(Conv1D_hist.history['val_acc'], label='Val accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
plt.title("Loss plot for Conv1D")
plt.plot(Conv1D_hist.history['loss'], label='Loss')
plt.plot(Conv1D_hist.history['val_loss'], label='Val loss')
plt.xlabel('Epochs')
plt.ylabel('Loss value')
plt.legend()
plt.show()
```

```python
# Convert test reviews to test sequences
test_sequences = tokenizer.texts_to_sequences(clean_reviews(test['review']))
test_data = pad_sequences(test_sequences, maxlen = max_length)

# Categorize test ratings 
test_ratings = test['rating'].values
test_labels = 1.0*(test_ratings >= 7) + 1.0*(test_ratings >= 5)
print(test_labels)
```

```python
# Apply 1D Convnet to test set
Conv1D_pred_labels = np.argmax(Conv1D_model.predict(test_data), axis = 1)
print(Conv1D_pred_labels)
```

```python
# Classification Report
Conv1D_cr = classification_report(test_labels, Conv1D_pred_labels)
print(Conv1D_cr)
```

```python
# Confusion Matrix
Conv1D_cm = confusion_matrix(test_labels, Conv1D_pred_labels).T
print(Conv1D_cm)
```

```python
# Accuracy from Grsser et al is 0.9224
# Cohen's kappa from Grsser et al is 0.8399

# 1D Convnet
print("Metrics for Conv1D:")
print("Accuracy                  =",classification_report(test_labels, Conv1D_pred_labels, output_dict=True)['accuracy'])
# Calculate the cohen's kappa, both with linear and quadratic weights for Conv1D
Conv1D_k = cohen_kappa_score(test_labels, Conv1D_pred_labels)
print(f"Cohen's kappa (linear)    = {Conv1D_k:.4f}")
Conv1D_k2 = cohen_kappa_score(test_labels, Conv1D_pred_labels, weights = 'quadratic')
print(f"Cohen's kappa (quadratic) = {Conv1D_k2:.4f}")
```
