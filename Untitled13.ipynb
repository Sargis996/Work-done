It is observed that the peak season is during the month 6 - 9. Based on the data analysis, high cancelation is driven by the booking deposit type even during the peak season especially month 7 & 8. However, it is viewed that the high cancelation rate does not affect the hotel business due to high adr and confirmed bookings during that period of time.

Referring to the guest types data, majority bookings are from new guests. Therefore, the bookings is viewed driven by the marketing, market segment & distribution channel in addition to the booking deposit types. This also provide the requirement for the hotel operator to improve promotion and engagement with existing or repeating guest.

The high online TA & offline TA & TO show significant contribution followed by the direct booking types. This shows that the direct distribution channel can also be the preferable booking and promotional channels to increase customer loyalty.

The high city hotel bookings are high compared to resort hotel. It is viewed that the bookings are driven by the commercial/business activities troughou the years based on the trend and bookings by coorporate & group. It is also assumed that the city hotel is located in or near the central business district (CBD). As for the resort hotel, though the adult number > 4 are considered outliers, it is viewed that the bookings with high adult numbers are driven by the corporate & group activities.


```python

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier
```

```python

df_proc['hotel'] = df_proc['hotel'].map({'Resort Hotel': 0, 'City Hotel': 1})
df_proc['meal'] = df_proc['meal'].map({'BB': 0, 'FB': 1, 'HC': 3, 'SC': 4})
df_proc['market_segment'] = df_proc['market_segment'].map({'Direct': 1, 'Corporate': 2, 'Online TA': 3, 'Offline TA/TO': 4, 'Complementary': 5,
                                                        'Groups': 6, 'Aviation':7})
df_proc['distribution_channel'] = df_proc['distribution_channel'].map({'Direct': 1, 'Corporate': 2, 'TA/TO': 3, 'GDS': 4})                                                        
df_proc['deposit_type'] = df_proc['deposit_type'].map({'No Deposit': 0, 'Refundable': 1, 'Non Refund': 3})
df_proc['customer_type'] = df_proc['customer_type'].map({'Transient': 1, 'Contract': 2, 'Transient-Party': 3, 'Group': 4})
```

```python
plt.figure(figsize=(14,14))

df_corr = df_proc.corr(method='spearman')
heatmap = sns.heatmap(df_corr.corr(), fmt=".2f", vmin=-1, vmax=1, annot=True)
plt.title("Correlation Matrix Spearman Method- Categorical Data ",size=15, weight='bold')
```

```python
df_corr = df_proc.corr(method = 'spearman')["is_canceled"]
df_corr.abs().sort_values(ascending=False)[1:]
```

```python
df_model = df_proc[['is_canceled', 'lead_time', 'required_car_parking_spaces', 'deposit_type', 'distribution_channel', 'adr', 'total_of_special_requests',
                'customer_type', 'previous_cancellations', 'booking_changes', 'total_nights', 'previous_bookings_not_canceled', 'stays_in_week_nights', 
                'company']]

label = 'is_canceled'
features = ['lead_time', 'required_car_parking_spaces', 'deposit_type', 'distribution_channel', 'adr', 'total_of_special_requests',
            'customer_type', 'previous_cancellations', 'booking_changes', 'total_nights', 'previous_bookings_not_canceled', 'stays_in_week_nights', 
            'company']
```

```python

X = df_model[features]
y = np.squeeze(df_model.drop(features,1))

X_train, X_test, y_train, y_test = train_test_split(
    X.values, y.values, test_size=0.2, random_state=123, shuffle = True)
```

```python
clf = RandomForestClassifier()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Accuracy Score of Random Forest is : {accuracy}")
print(f"Confusion Matrix : \n{conf_matrix}")
print(f"Classification Report : \n{report}")
```
